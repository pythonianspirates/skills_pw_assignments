#### Ques.1 - How can you create a Bokeh plot using Python code?

```python
from bokeh.plotting import figure, show 
import bokeh.io

bokeh.io.output_notebook()

x = [4,2,8,3,7,1]
y = [3,5,1,9,4,6]

p = figure(title="Bokeh Plot")
p.circle(x,y)
show(p)
```

#### Ques.2 - What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

In Bokeh, glyphs are visual markers or symbols that you can add to a plot to represent data points. They can be used to display different types of information, such as points, lines, bars, or areas on a graph. Adding glyphs to a Bokeh plot helps to make your data more visually appealing and easier to understand.

To add glyphs to a Bokeh plot, you need to use the **'figure'** class from the Bokeh library.

```python
from bokeh.plotting import figure ,show 

# create a new figure 
p = figure(title='Scatter Plot Example',x_axis_label='X-axis',y_axis_label='Y-axis')

#Define some Datapoint
x = [4,2,8,3,7,1]
y = [3,5,1,9,4,6]

#Add Circles as glyphs to the plot 
p.circle(x,y,size=10,color='red',alpha = 0.8)

# Show the plot 
show(p)
```

In this example, we first create a new figure using the figure class and specify a title for the plot along with labels for the x-axis and y-axis. Then, we define the data points that we want to plot using the x and y lists.

To add circles as glyphs to the plot, we use the circle method of the figure object. We pass the x and y data as the first two arguments, followed by other optional parameters like size (to specify the size of the circles), color (to set the color of the circles), and alpha (to control the transparency of the circles).

Finally, we use the show function to display the plot. When you run this code, you'll see a scatter plot with circles representing the data points specified by the x and y lists.

#### Ques.3 -How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

We can customize the appearance of a Bokeh plot including the axes,title,and the legend by using various parameters and methods provided by Bokeh library.

For example - 

```python
from bokeh.plotting import figure,show
import bokeh.io

bokeh.io.output_notebook()

#create some data
x = [1,2,3,4,5]
y = [6,5,4,3,2]

# create a new plot with customized appearance 
p = figure(title="bokeh plot")

# add circle glyphs to the plot with  Legend Label
p.circle(x,y,legend_label="My Data",size=10,color='black')

#customize the x-axis and y-axis labels and ticks
p.xaxis.axis_label = "X-axis label"
p.yaxis.axis_label = "Y-axis label"

# show the plot with legend in top-left position
p.legend.location = "top_left"
show(p)
```

#### Ques.5 -How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?


To embed a Bokeh plot into a web page or dashboard using Flask or Django:

- Create your Bokeh plot.
- Save the plot as an HTML file using Bokeh's save() function.
- Set up your Flask or Django application.
- Create a route or view to handle the web page or dashboard where the plot will be embedded.
- Render the HTML file in the template or view to embed the Bokeh plot.
- Start your Flask or Django application to serve the web page or dashboard with the embedded Bokeh plot.
