### Ques.2 - What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
        import numpy as np
        np.random.seed(3)
        x = 3 + np.random.normal(0, 2, 50)
        y = 3 + np.random.normal(0, 2, len(x))
#### Note: Also add title, xlabel, and ylabel to the plot.

- A scatter plot is a type of data visualization that displays a collection of data points as individual markers on a 2D coordinate system, with one variable on the x-axis and another variable on the y-axis. Each point represents the values of two variables for a single observation.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y,c=x)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of x and y')
plt.show()
```

### Ques.3 - Why is the subplot() function used? Draw four line plots using the subplot() function.
        Use the following data:
        import numpy as np

        y = np.array([0, 100, 200, 300, 400, 500])
        for line 2: 
        x = np.array([0, 1, 2, 3, 4, 5]) and 
        y = np.array([50, 20, 40, 20, 60, 70])
        for line 3: 
        x = np.array([0, 1, 2, 3, 4, 5]) and 
        y = np.array([10, 20, 30, 40, 50, 60])
        for line 4:
        x = np.array([0, 1, 2, 3, 4, 5]) and 
        y = np.array([200, 350, 250, 550, 450, 150])

- The subplot() function in Python is used to create multiple plots in the same figure. It allows you to divide the figure into a grid of rows and columns, and select a specific cell in which to plot your data.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5]) 
    
y0 = np.array([0, 100, 200, 300, 400, 500])

y1 = np.array([50, 20, 40, 20, 60, 70])
    
y2 = np.array([10, 20, 30, 40, 50, 60])
    
y3 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,1)
plt.plot(x,y0)
plt.title("First LP")

plt.subplot(2,2,2)
plt.plot(x,y1)
plt.title("Second LP")

plt.subplot(2,2,3)
plt.plot(x,y2)
plt.title("Third LP")

plt.subplot(2,2,4)
plt.plot(x,y3)
plt.title("Fourth LP")
```

### Ques.4 - What is a bar plot? Why is it used? Using the following data, plot a bar plot and a horizontal bar plot.
        import numpy as np
        company = np.array(["Apple", "Microsoft","Google", "AMD"])
        profit = np.array([3000, 8000, 1000, 10000])

- A bar plot is a graphical representation of data using bars of different heights or lengths to represent the values of the variables being plotted. It is commonly used to compare and visualize the distribution, frequency or proportion of categorical or discrete data.

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft","Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

#Bar Plot
plt.bar(company,profit)
plt.title('Company V/S Profit')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.show()


#Horizontal Bar Plot
plt.barh(company,profit)
plt.title('Company V/S Profit')
plt.ylabel('Company')
plt.xlabel('Profit')
plt.show()
```

### Ques.5 - What is a box plot? Why is it used? Using the following data plot a box plot.
                box1 = np.random.normal(100, 10, 200)
                box2 = np.random.normal(90, 20, 200)

-  A box plot is a graphical representation of the distribution of numerical data through their quartiles, median and outliers. It is used to display and compare the spread, skewness, and symmetry of different datasets or variables, and to identify potential outliers or extreme values.

```python
import numpy as np 
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

pw =[box1,box2]
fig,ax =plt.subplots()
ax.boxplot(pw)
plt.show()
```
